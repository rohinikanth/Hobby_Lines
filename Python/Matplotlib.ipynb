```python
import matplotlib.pyplot as plt

month =["jan","feb","marc","apr","may","jun","jul"]
temp=[28,35,35,32,34,27,30]

plt.figure(figsize=(10,5))
plt.plot(month,temp,marker="o",linestyle = ":",linewidth=2,color='red')
plt.title('Average monthly temperature')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temperature = [25, 27, 30, 33, 35, 34, 32, 30, 39, 28, 26, 25]
rainfall = [12, 18, 25, 45, 120, 220, 200, 150, 80, 30, 15, 10]
humidity = [60, 65, 70, 75, 85, 90, 92, 88, 80, 70, 65, 60]
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig1 , ax1 =plt.subplots()
ax1.plot(months,rainfall,color='green',marker='')
ax1.set_title("Month v/s Rainfall")
ax1.set_xlabel("Month")
ax1.set_ylabel("mm")


fig2 , ax2 =plt.subplots()
ax2.scatter(months,rainfall,s=10)
ax2.set_title("Monthly Rainfall - Scatter Plot")
ax2.set_xlabel("Month")
ax2.set_ylabel("mm")

fig3, ax3 =plt.subplots()
ax3.bar(months,rainfall)
ax3.set_title("Monthly Rainfall - Bar Plot")
ax3.set_xlabel("Month")
ax3.set_ylabel("mm")


plt.show()
```

bbox_to_anchor=(0, 1, 1, 0.1)	Places the legend in a custom box: (x0, y0, width, height) relative to the axes

loc="lower left"	Aligns the legend’s "lower left" corner to the anchor box

mode="expand"	Expands the legend entries to fill the width of the box

ncol=2	Displays the legend in 2 columns

```python
fig4, ax4 = plt.subplots()
ax4.hist(temperature, bins=7, color='lime', edgecolor='black')
ax4.set_title("Temperature Distribution(Histogram)")  #Frequency of temp values

ax4.set_title("Temperature Distribution")
ax4.set_xlabel("Temperature (°C)")
ax4.set_ylabel("Number of Months")

ax4.legend(bbox_to_anchor=(0,1,1,.3), ncol=4,mode="expand", loc="right")
plt.show()
```
